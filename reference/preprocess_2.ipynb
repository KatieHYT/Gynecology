```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_n_zeros(d):
    n_zeros = list()
    for i in range(d.shape[0]):
        n_zeros.append(sum(d[i,:] ==0))
    return np.array(n_zeros)
```

```python
d_cmu_b = d_cmu[[k for k in d_cmu.columns if 'b-' in k]]
d_cmu_m = d_cmu[[k for k in d_cmu.columns if 'm-' in k]]
d_cmu_l = d_cmu[[k for k in d_cmu.columns if ('m-' not in k and 'b-' not in k)]]
print(d_cmu.shape)
print(d_cmu_b.shape)
print(d_cmu_m.shape)
print(d_cmu_l.shape)
```

```python
d_ctu_b = d_ctu[[k for k in d_ctu.columns if 'b-' in k]]
d_ctu_m = d_ctu[[k for k in d_ctu.columns if 'm-' in k]]
d_ctu_l = d_ctu[[k for k in d_ctu.columns if ('m-' not in k and 'b-' not in k)]]
print(d_ctu.shape)
print(d_ctu_b.shape)
print(d_ctu_m.shape)
print(d_ctu_l.shape)
```

```python
d_cmu_l['ID'] = 'CMU_'+d_cmu_l['ID'].astype('str')+'_'+d_cmu_l['seconds'].astype('str')
d_ctu_l['ID'] = 'CTU_'+d_ctu_l['ID'].astype('str')+'_'+d_ctu_l['time'].astype('str')
col_list = ['ID', 'management', 'deceleration', 'variability','FHB' ,'UA', 'UA_duration', 'UA_interval']
```

```python
d_cmu_l_keep = d_cmu_l[col_list+['datetime','seconds', 'bed', 'start', 'end']]
d_ctu_l_keep = d_ctu_l[col_list+['pH', 'BDecf', 'pCO2', 'BE', 'Apgar1', 'Apgar5']]
```

```python
d_cmu_l_keep.shape
```

```python
d_ctu_l_keep.shape
```

```python
d_merge = pd.concat([d_cmu_l_keep,d_ctu_l_keep],axis=0,ignore_index=True)
d_merge = d_merge[col_list+['datetime','seconds', 'bed', 'start', 'end', 'pH', 'BDecf', 'pCO2', 'BE', 'Apgar1', 'Apgar5']]
```

```python
d_ctu_b_mani = d_ctu_b.rolling(4, axis = 1).mean() 
d_ctu_b_mani = d_ctu_b_mani.iloc[:, ::-1]
d_ctu_b_mani = d_ctu_b_mani.iloc[:, ::4]
d_ctu_b_mani = d_ctu_b_mani.iloc[:, ::-1]

d_ctu_m_mani = d_ctu_m.rolling(4, axis = 1).mean() 
d_ctu_m_mani = d_ctu_m_mani.iloc[:, ::-1]
d_ctu_m_mani = d_ctu_m_mani.iloc[:, ::4]
d_ctu_m_mani = d_ctu_m_mani.iloc[:, ::-1]
```

```python
a = np.array(d_cmu_b)
b = np.array(d_ctu_b_mani)
b_merge = pd.DataFrame(np.concatenate((a, b), axis=0), columns=d_cmu_b.columns)


a = np.array(d_cmu_m)
b = np.array(d_ctu_m_mani)
m_merge = pd.DataFrame(np.concatenate((a, b), axis=0), columns=d_cmu_m.columns)
```

```python
dbm_merge = pd.concat([d_merge, b_merge, m_merge], axis = 1)
```

```python
dbm_merge.to_csv(save_to,index=False)
```
